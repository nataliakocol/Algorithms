```python
#quicksort
def m_i(a, b, c):
    if (a <= b and c <= a) or (a <= c and b <= a):
        return 0
    elif (b <= a and c <= b) or (b <= c and a <= b):
        return 1
    else:
        return 2
    
def get_p_i(a, lo, hi):
    i = [random.randint(lo, hi) for _ in range(3)]
    m = m_i(a[i[0]], a[i[1]], a[i[2]])
    return i[m]
    
def partition(a, lo, hi):
    p_i = get_p_i(a, lo, hi)
    a[lo], a[p_i] = a[p_i], a[lo]
    p = a[lo]
    
    border = lo
    for i in range(lo, hi + 1):
        if a[i] < p:
            border += 1
            a[i], a[border] = a[border], a[i]
            
    a[lo], a[border] = a[border], a[lo]
    return border

def sort(a, lo, hi):
    if lo < hi:
        p = partition(a, lo, hi)
        sort(a, lo, p - 1)
        sort(a, p + 1, hi)
        
def q_sort(a):
    sort(a, 0, len(a) - 1)
```

```python
#binary heap
h = [-1]

def enqueue(key):
    h.append(key)
    swim(len(h) - 1)
    
def swim(i):
    while i > 1 and h[i//2] < h[i]:
        h[i//2], h[i] = h[i], h[i//2]
        i = i//2
        
def dequeue():
    maxi = h[1]
    h[1], h[len(h) - 1] = h[len(h) - 1], h[1]
    del h[-1] #or h.pop()
    sink(1)
    
    return maxi

def sink(i):
    while 2*i <= len(h) - 1:
        j = 2*i
        if j < len(h) - 1 and h[j] < h[j + 1]:
            j += 1
        if h[i] >= h[j]:
            break
        h[i], h[j] = h[j], h[i]
        i = j
```

```python
#heap sort
#we are given an array in random order, a
#since a starts with elements at 0, children are 2i+1 and 2i+2

#heap construction (bottom up):
def sink_2(a, i, N):
    while 2*i + 1 < N:
        j = 2*i + 1
        if j < N - 1 and a[j] < a[j + 1]:
            j += 1
        if a[i] >= a[j]:
            break
        a[i], a[j] = a[j], a[i]
        i = j
    
def heapify(a):
    for k in range(len(a)//2, -1, -1):
        sink_2(a, k, len(a))

def sortdown(a):
    N = len(a) - 1
    while N > 0:
        a[0], a[N] = a[N], a[0]
        N -=1
        sink_2(a, 0, N + 1)
        
def heap_sort(a):
    heapify(a)
    sortdown(a)
```

```python
#copying for lab exc.
#merge step
def merge(arr, aux, lo, mid, hi):
    i = lo
    j = mid + 1
    for k in range(lo, hi+1):
        if i > mid:
            arr[k] = aux[j]
            j += 1
        elif j > hi:
            arr[k] = aux[i] 
            i += 1
        elif aux[i] < aux[j]:
            arr[k] = aux[i]
            i += 1
        else:
            arr[k] = aux[j]
            j += 1
            
#merge sort
def sort(arr, aux, lo, hi):
    if hi <= lo:
        return
    mid = lo + ((hi - lo)//2)
    sort(aux, arr, lo, mid)
    sort(aux, arr, mid + 1, hi)
    merge(arr, aux, lo, mid, hi)

def merge_sort(arr):
    aux = arr.copy()
    sort(arr, aux, 0, len(arr) - 1)
```

```python
#LABS
#triplicates: given 3 arrays of n strings, determine 
#if there is a common string to all 3 of them
#let's try to use merge & heap sorts

#ver 1:merge sort
import time
import random
import string
a = [''.join(random.choice(string.ascii_lowercase) for i in range(random.randint(1,10))) for i in range (10)]
b = [''.join(random.choice(string.ascii_lowercase) for i in range(random.randint(1,10))) for i in range (10)]
c = [''.join(random.choice(string.ascii_lowercase) for i in range(random.randint(1,10))) for i in range (10)]
a.append('ad')
b.append('ad')
c.append('ad')

def bin_search(x, a):
    low = 0
    high = len(a) - 1
    while low <= high:
        mid = low + int((high - low)/2)
        if a[mid] < x: 
            low = mid + 1
        elif a[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1

def triplicates_1(a, b, c):
    a_1 = a.copy()
    b_1 = b.copy()
    c_1 = c.copy()
    #we do not need to sort a_1 actually merge_sort(a_1)
    merge_sort(b_1)
    merge_sort(c_1)
    for x in a_1:
        if bin_search(x, b_1) != -1:
            found = bin_search(x, c_1)
            if found != -1:
                return found
    return "not found"
    
#ver 2:heap sort
def triplicates_2(a, b, c):
    a_1 = a.copy()
    b_1 = b.copy()
    c_1 = c.copy()
    #heap_sort(a_1)
    heap_sort(b_1)
    heap_sort(c_1)
    for x in a_1:
        if bin_search(x, b_1) != -1:
            found = bin_search(x, c_1)
            if found != -1:
                return found
    return "not found"

#ver 3:merge into one array, sort, find triplicates
def triplicates_3(a, b, c):
    arr = a + b + c
    heap_sort(arr)
    i = 0
    while i < len(arr) - 3:
        if arr[i + 1] == arr[i]:
            if arr[i + 2] == arr[i]:
                return i
            else:
                i += 2
        else:
            i += 1
    print(arr[i])

print(a, b, c)

start = time.time()
print("first method")
triplicates_1(a, b, c)
end = time.time()
print(end-start)

start = time.time()
print("second method")
triplicates_2(a, b, c)
end = time.time()
print(end-start)

start = time.time()
print("third method")
print(triplicates_3(a, b, c))
end = time.time()
print(end-start)
```

```python
#ex 4
#we assume jobs can overlap & start and end times 
#are given in pairs (tuples) for ex. 1230, 1200, 1500
#if a start time, add in tuple as (time, "s")
def process_times(a):
    b = []
    for job in a:
        b.append((job[0],"s"))
        b.append((job[0],"e"))
    merge_sort(b)
    return b
        
def jobs(a):
    a = process_times(a)
    busy = idle = start = end = count = 0
    for j in a:
        if j[1] == "s":
            if count == 0:
                start = j[0]
                idle = max(idle, start - end)
            count += 1
        else:
            count -= 1
            if count == 0:
                end = j[0]
                busy = max(busy, end - start)
```

```python
#lab 4
#1: pancake sorting
import random
#pancakes = random.sample(range(100), 10)
pancakes = [3, 2, 4, 1]

def flip(i):
    start = 0
    while start < i:
        pancakes[start], pancakes[i] = pancakes[i], pancakes[start]
        start += 1
        i -= 1
    
        
def pancake_sort():
    indices = []
    N = len(pancakes)
    count = 0
    for i in range(N - 1, 0, -1):
        max_ = i
        for j in range(i-1, -1, -1):
            if pancakes[j] > pancakes[max_]:
                max_ = j
        print(max_)
        if max_ != i:
            flip(max_)
            flip(i)
            indices.append(max_+1)
            indices.append(i+1)
            count += 2
    return (indices, count)

#improving? sort by finding each of i from 1 to i, and sorting in reverse order, then flipping all to top
def improved_pancake():
    N = len(pancakes)
    current = 1
    count = 0
    indices = []
    for i in range(N - 1, -1, -1):
        min_ = i
        while pancakes[min_] != current:
            min_ -= 1
        if min_ != i:
            flip(min_)
            flip(i)
            count += 2
            indices.append(min_)
            indices.append(i)
        current += 1
    count += 1
    flip(N - 1)
    indices.append(N-1)
    return (count, indices)
            
print(improved_pancake())
print(pancakes)
```
