```python
#ex1
class Node:
    def __init__(self, value):
            self.value = value
            self.next = None

class CircularList:
    def __init__(self):
        self.head = self.tail = None
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def length(self):
        return self.size
    
    def append(self, item):
        if self.isEmpty():
            self.head = self.tail = item
            item.next = self.head
        else:
            self.tail.next = item
            item.next = self.head
            self.tail = item
        self.size +=1
        
    def prepend(self, item):
        if self.isEmpty():
            self.head = self.tail = item
            item.next = self.head
        else:
            self.tail.next = item
            item.next = self.head
            self.head = item
        self.size +=1
        
    def delete(self, pos):
        if self.isEmpty() or pos < 0 or pos >= self.size:
            print("Error")
            return None
        elif pos == 0:
            item = self.head
            self.tail.next = item.next
            self.head = item.next
            self.size -=1
            return item
        elif pos == self.size-1:
            item = self.tail
            self.tail = self.access(pos - 1)
            self.tail.next = self.head
            self.size -= 1
            return item
 
        prev = self.access(pos - 1)
        item = prev.next
        prev.next = item.next
        self.size -=1
        return item
    
    def access(self, pos):
        if pos < 0 or pos >= self.size:
            print("Error")
            return None
        elif pos == self.size - 1:
            return self.tail
        
        i = 0
        cur = self.head
        while i < pos:
            cur = cur.next
            i += 1
        
        return cur
    
```

```python
#ex2
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class Deque:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def isEmpty(self):
        return self.size == 0
    
    def length(self):
        return self.size
    
    def addFirst(self, item):
        self.size += 1
        if self.size == 0:
            self.head = self.tail = item
        else:
            item.next = self.head
            self.head = item
        
    def addLast(self, item):
        self.size += 1
        if self.size == 0:
            self.head = self.tail = item
        else:
            item.prev = self.tail
            self.tail = item
            
    def removeFirst(self):
        if self.size == 0:
            print("Error")
            return None
        self.size -= 1
        item = self.head
        self.head = item.next
        self.head.prev = None
        return item
    
    def removeLast(self):
        if self.size == 0:
            print("Error")
            return None
        self.size -= 1
        item = self.tail
        self.tail = item.prev
        self.tail.next = None
        return item
```

```python
#ex3
def palindromeChecker(text):
    if len(text) == 0:
        print("Error: no text entered")
        return False
    
    text = text.lower() 
    start = 0
    end = len(text) - 1
    while start <= end:
        if text[start]!=text[end]:
            return False
        start += 1
        end -= 1
    return True
```

```python
#LAB exercises
#ex1: unique paths
r, c = map(int, input("Enter number of rows and columns:").split())

if r < 2 or c < 2:
    print("Number of rows/columns needs to be equal to or greater than 2") 

dp = [[0 for i in range(c + 1)] for j in range(r + 1)]
#corner cases
dp[2][2] = 2
for i in range(3, r + 1):
    dp[i][2] = dp[i - 1][2] + 1
for j in range(3, c + 1):
    dp[2][j] = dp[2][j - 1] + 1

for i in range(3, r + 1):
    for j in range(3, c + 1):
        dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

print("The number of unique paths is equal to:", dp[r][c])
    
```

```python
#ex3
def middle(a):
    low = 0
    high = len(a) - 1
    while low <= high:
        mid = low + int((high - low)/2)
        #do we need to handle errors..?
        if a[mid - 1] < a[mid] and a[mid] < a[mid + 1]:  
            low = mid + 1
        elif a[mid - 1] > a[mid] and a[mid] > a[mid + 1]:
            high = mid - 1
        else:
            return mid
    return -1

def bin_up(x, a, middle):
    low = 0
    high = middle
    while low <= high:
        mid = low + int((high - low)/2)
        if a[mid] < x: 
            low = mid + 1
        elif a[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1

def bin_down(x, a, middle):
    low = middle
    high = len(a) - 1
    while low <= high:
        mid = low + int((high - low)/2)
        if a[mid] < x: 
            high = mid - 1
        elif a[mid] > x:
            low = mid + 1
        else:
            return mid
    return -1

a = list(map(int, input("Enter the array elements:").split()))
x = int(input("Enter the element you are looking for:"))
m = middle(a)         
index = max(bin_up(x, a, m), bin_down(x, a, m))

if index == -1:
    print("Element not found")
else:
    print("Element found at index", index)
        
```
